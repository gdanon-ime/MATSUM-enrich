```python
lucas = ee.FeatureCollection('JRC/LUCAS_HARMO/THLOC/V1')\
  .select(['point_id', 'lu1'])\
    .limit(1000)

# reclassify lucas data lc names
df = pd.read_csv('lucas_legends.csv', sep=';', error_bad_lines=False)
df['code'] = df['code'].str.strip("'").astype(str)

df_filter = df[(df['var'] == 'lu1') | (df['var'] == 'lu1_type')]

df_code = ee.List(df_filter['code'].tolist())
df_label = ee.List(df_filter['label'].tolist())

# Map over the 'lucas' Feature Collection
lucas = lucas.map(lambda ft: ft.set('lu1Name', df_label.get(
    df_code.indexOf(ft.get('lu1')))))

lucas_df = geemap.ee_to_pandas(lucas)
lucas_df
```
